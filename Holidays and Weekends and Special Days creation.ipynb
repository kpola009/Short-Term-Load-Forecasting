```python
weekends_full_frames = [weekends_1987, weekends_1988, weekends_1989]
weekends_full = pd.concat(weekends_full_frames)
```

```python
weekends_full = weekends_full.reset_index()
weekends_full.drop(columns = ['index'], inplace = True)
```

```python
holidays_full_frames = [holidays_1987, holidays_1988, holidays_1989]
holidays_full = pd.concat(holidays_full_frames)
```

```python
holidays_full = holidays_full.reset_index()
holidays_full.drop(columns = ['index'], inplace = True)
```

```python
special_day_full_frames = [special_day_1987, special_day_1988, special_day_1989]
special_day_full = pd.concat(special_day_full_frames)
```

```python
special_day_full = special_day_full.reset_index()
special_day_full.drop(columns = ['index'], inplace = True)
```

```python
def df_preprocess(df):
    df = df.drop(columns=['Year', 'Day Number', 'Day Type', 'Holiday', 'Season', 'Weekday', 'Weekend', 'Special day'], 
            axis = 1)
    
    arr = df.to_numpy()
    rows = arr.shape[0] * arr.shape[1]
    
    arr = arr.reshape((rows, 1))
    
    updated_df = pd.DataFrame(arr, columns = ['Load'])
    
    final_df = pd.concat([updated_df], axis = 1)  
    return final_df
```

```python
updated_1987 = df_preprocess(year_1987)
updated_1988 = df_preprocess(year_1988)
updated_1989 = df_preprocess(year_1989)
```

```python
train = [updated_1987, updated_1988, updated_1989]

train_df = pd.concat(train)
```

```python
train_df = train_df.reset_index()
train_df
```

#### Create Datetime column

```python
datetime_df = pd.DataFrame({
    'date':pd.date_range(start='1987-01-01', end='1990-01-01', freq = 'H')
})
```

```python
datetime_df.drop(datetime_df.tail(1).index, inplace = True)
datetime_df
```

```python
train_df = pd.concat([datetime_df, train_df], axis=1)
```

```python
train_df.drop(columns = ['index'], inplace = True)
train_df
```

```python
final_train_df = pd.concat([train_df, weekends_full, holidays_full, special_day_full], axis = 1)
final_train_df
```
